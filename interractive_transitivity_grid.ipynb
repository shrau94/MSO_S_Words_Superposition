```python
from tkinter import *
from tkinter import font as tkFont
import math
import  tkinter as tk

'''
Superposition implementation on the given two strings
'''
def superpose(string1, string2, voc1, voc2):
    if len(string1) == len(string2) == 0:
        return [string1]
    else:
        temp = []
        if len(string1) > 0:
            h1 = string1[0]
            t1 = string1[1:]
            h1v2 = h1.intersection(voc2)
            if len(h1v2) == 0:
                for res in superpose(t1,string2,voc1,voc2):
                    temp.append([h1] + res)     
            if len(string2) > 0:
                h2 = string2[0]
                t2 = string2[1:]
                h3 = h1.union(h2)
                if h1v2.issubset(h2):
                    h2v1 = h2.intersection(voc1)
                    if h2v1.issubset(h1):
                        for res in superpose(t1,t2,voc1,voc2):
                            temp.append([h3] + res)
        if len(string2) > 0:
            h2 = string2[0]
            t2 = string2[1:]
            h2v1 = h2.intersection(voc1)
            if len(h2v1) == 0:
                for res in superpose(string1,t2,voc1,voc2):
                    temp.append([h2] + res)
        return temp

def super(s1,s2):
    return superpose(s1,s2,voc(s1),voc(s2))

def voc(string): 
    if len(string) == 0:
        return {}
    else:
        return string[0].union(voc(string[1:]))

def proj(string,set): 
    if len(string) == 0:
        return string
    else:
        h1 = string[0].intersection(set)
        if len(h1) == 0:
            return proj(string[1:],set)
        else:
            return [h1] + proj(string[1:],set)

def l(x):
    return "l" + str(x)

def r(x):
    return "r" + str(x)

def allen(x,y,rel):
    lx = l(x)
    rx = r(x)
    ly = l(y)
    ry = r(y)
    if rel == "b":
        return [{lx}, {rx}, {ly}, {ry}]
    elif rel == "o":
        return [{lx}, {ly}, {rx}, {ry}]
    elif rel == "m":
        return [{lx}, {rx,ly}, {ry}]
    elif rel == "d":
        return [{ly}, {lx}, {rx}, {ry}]
    elif rel == "s":
        return [{lx,ly}, {rx}, {ry}]
    elif rel == "f":
        return [{ly}, {lx}, {rx,ry}]
    if rel == "bi":
        return [{ly}, {ry}, {lx}, {rx}]
    elif rel == "oi":
        return [{ly}, {lx}, {ry}, {rx}]
    elif rel == "mi":
        return [{ly}, {ry,lx},{rx}]
    elif rel == "di":
        return [{lx}, {ly}, {ry}, {rx}]
    elif rel == "si":
        return [{lx,ly}, {ry}, {rx}]
    elif rel == "fi":
        return [{lx}, {ly}, {rx,ry}]
    elif rel == "eq":
        return [{lx,ly},{rx,ry}]    
    
def allInv(string,s1,s2):
    lx = l(s1)
    rx = r(s1)
    ly = l(s2)
    ry = r(s2)
    if string == [{lx}, {rx}, {ly}, {ry}]:
        return "b"
    elif string == [{lx}, {ly}, {rx}, {ry}]:
        return "o"
    elif string == [{lx}, {rx,ly}, {ry}]:
        return "m"
    elif string == [{ly}, {lx}, {rx}, {ry}]:
        return "d"
    elif string == [{lx,ly}, {rx}, {ry}]:
        return "s"
    elif string == [{ly}, {lx}, {rx,ry}]:
        return "f"
    elif string == [{ly}, {ry}, {lx}, {rx}]:
        return "bi"
    elif string == [{ly}, {lx}, {ry}, {rx}]:
        return "oi"
    elif string == [{ly}, {ry,lx},{rx}]:
        return "mi"
    elif string == [{lx}, {ly}, {ry}, {rx}]:
        return "di"
    elif string ==  [{lx,ly}, {ry}, {rx}]:
        return "si"
    elif string == [{lx}, {ly}, {rx,ry}]:
        return "fi"
    elif string == [{lx,ly},{rx,ry}]:
        return "eq"

'''
Computes the transitive relations for the given two relations
'''
def transitivity_relation(r1,r2):
    temp = []
    superpose_temp=[]
    for s in super(allen(0,1,r1), allen(1,2,r2)):
        superpose_temp.append(s)
        temp.append(allInv(proj(s,{l(0),r(0),l(2),r(2)}), 0, 2))
    return temp,superpose_temp

'''
Computations all ranked strings
'''
def rank0():
    '''
    Allen's 13 relations
    '''
    return ['b','bi','d','di','eq','f','fi','m','mi','o','oi','s','si']

def rank1():
    rank1_strings = []
    rank0_strings = rank0()
    rank1_dictionary = []
    for r1 in rank0_strings:
        for r2 in rank0_strings:
            temp,s = transitivity_relation(r1,r2)
            if temp not in rank1_strings:
                if len(temp) != 1:
                    temp.sort()
                    if not is_present(temp,rank1_strings):
                        rank1_dictionary.append({
                        "relation" : (r1, r2),
                        'string': temp
                        })
                        rank1_strings.append(temp)
    return rank1_strings,rank1_dictionary

def rank2():
    rank0_strings = rank0()
    rank1_strings,rank1_dictionary = rank1()
    rank2_strings = []
    rank2_dictionary = []
    for r1s in rank1_strings:
        for r2s in rank1_strings:
            temp = []
            for r1 in r1s:
                for r2 in r2s:
                    relation,s = transitivity_relation(r1,r2)
                    temp.extend(relation)
            temp.sort()
            temp2 = list(set(temp)) 
            temp2.sort()
            if not is_present(temp2,rank2_strings) and not is_present(temp2,rank1_strings) and not is_present(temp2,rank0_strings):
                if len(temp) != 1:
                    rank2_dictionary.append({
                        'relation' : (r1s, r2s),
                        'string': temp2
                    })
                    rank2_strings.append(temp2)
    return rank2_strings,rank2_dictionary

def rank3():
    rank0_strings = rank0()
    rank1_strings,rank1_dictionary = rank1()
    rank2_strings,rank2_dictionary = rank2()
    rank3_strings = []
    for r1s in rank2_strings:
        for r2s in rank2_strings:
            temp = []
            for r1 in r1s:
                for r2 in r2s:
                    relation,s = transitivity_relation(r1,r2)
                    temp.extend(relation)
            temp.sort()
            temp2 = list(set(temp))
            temp2.sort()
            if not is_present(temp2,rank3_strings) and not is_present(temp2,rank2_strings) and not is_present(temp2,rank1_strings) and not is_present(temp2,rank0_strings):
                if len(temp) != 1:
                    rank3_strings.append(temp2)
    return rank3_strings

'''
Checks if a relation is present in the list
'''
def is_present(item, rank_list):
    is_present = False
    for rank_item in rank_list:
        if(set(rank_item) == set(item)):
            is_present = True
            break;
    return is_present

'''
Computes the transitivity table for all the allen relations
'''
def compute_transitivty_table():
    transitivity_array = []
    rowNumber = 0
    for row in range(14):
        columnNumber = 0
        rowNumber = rowNumber + 1
        temp = []
        super_temp=[]
        for col in range(14):
            columnNumber = columnNumber + 1
            if rowNumber == 1 and columnNumber == 1:
                temp.append("Allen \nRelations")
                super_temp.append("Allen \nRelations")
            elif rowNumber == 1 and columnNumber != 1:
                temp.append(all_relations[columnNumber-2])
                super_temp.append(all_relations[columnNumber-2])
            elif columnNumber == 1 and rowNumber != 1:
                temp.append(all_relations[rowNumber-2])
                super_temp.append(all_relations[rowNumber-2])
            else:
                temp1,supertemp1=transitivity_relation(all_relations[rowNumber-2], all_relations[columnNumber-2])
                temp1.sort()
                temp.append(temp1)
                super_temp.append(supertemp1)
        superpose_result.append(super_temp)
        transitivity_array.append(temp)
    return transitivity_array 
   
'''
Computes the possible short string representation for all relations
'''
def possible_s_words():
    global possible_short_representations
    all_allen_possibilities = super([{'li'}, {'ri'}],[{'lj'}, {'rj'}])
    allen_for_representation = []
    for rep in possible_short_representations:
        temp = []
        for allen in all_allen_possibilities:
            for s in super(allen, rep):
                temp.append(allInv(proj(s,{l('i'),r('i'),l('j'),r('j')}), 'i', 'j'))
        allen_for_representation.append(temp)
    return allen_for_representation
 
'''
Gets the possible short string representation for a relation
'''       
def get_shorter_s_word(relation_array):
    global shorter_representations, possible_short_representations
    i = 0
    while(i<len(shorter_representations)):
        if(set(relation_array) == set(shorter_representations[i])):
            return possible_short_representations[i]
        i = i+1
    
'''
Drawing the transitivity grid on the canvas along with the text inside each box
'''
def draw_grid(view, selected, selected1, transitive1, transitive2):
    w=1400
    h=600
    gridWidth = w / 14
    gridHeight = h / 14
    rowNumber = 0
    selected_row, selected_column = selected
    selected_row1, selected_column1 = selected1
    transitive_row1, transitive_column1 = transitive1
    transitive_row2, transitive_column2 = transitive2
    for row in range(14):
        columnNumber = 0
        rowNumber = rowNumber + 1
        for col in range(14):
            columnNumber = columnNumber + 1
            if rowNumber == 1 and columnNumber == 1:
                rect = view.create_rectangle(col * gridWidth,row * gridHeight,(col + 1) * gridWidth,(row + 1) * gridHeight,fill ='light grey',outline='white')
                rect_text= transitivity_array[rowNumber-1][columnNumber-1]
            elif rowNumber == 1 and columnNumber != 1:
                rect = view.create_rectangle(col * gridWidth,row * gridHeight,(col + 1) * gridWidth,(row + 1) * gridHeight,fill ='light grey',outline='white')
                rect_text= transitivity_array[rowNumber-1][columnNumber-1]
            elif columnNumber == 1 and rowNumber != 1:
                rect = view.create_rectangle(col * gridWidth,row * gridHeight,(col + 1) * gridWidth,(row + 1) * gridHeight,fill ='light grey',outline='white')
                rect_text= transitivity_array[rowNumber-1][columnNumber-1]
            elif (transitive_row1 == rowNumber and transitive_column1 == columnNumber) or (transitive_row2 == rowNumber and transitive_column2 == columnNumber):
                rect = view.create_rectangle(col * gridWidth,row * gridHeight,(col + 1) * gridWidth,(row + 1) * gridHeight,fill = 'tomato',outline='white',width=2)
                rect_text= transitivity_array[rowNumber-1][columnNumber-1]
                if len(list(rect_text)) == 13:
                    rect_text = "All \nrelations"
                elif(len(list(rect_text)) > 4):
                    rect_text = (str(list(rect_text)[0:4]) + '\n' + str(list(rect_text)[4:])).replace(',','').replace('\'','').replace('[','').replace(']','')
            elif (selected_row == rowNumber and selected_column == columnNumber) or (selected_row1 == rowNumber and selected_column1 == columnNumber):
                rect = view.create_rectangle(col * gridWidth,row * gridHeight,(col + 1) * gridWidth,(row + 1) * gridHeight,fill = 'medium spring green',outline='black',width=2)
                rect_text= transitivity_array[rowNumber-1][columnNumber-1]
                if len(list(rect_text)) == 13:
                    rect_text = "All \nrelations"
                elif(len(list(rect_text)) > 4):
                    rect_text = (str(list(rect_text)[0:4]) + '\n' + str(list(rect_text)[4:])).replace(',','').replace('\'','').replace('[','').replace(']','')
            else:
                if len(transitivity_array[rowNumber-1][columnNumber-1]) == 1 and transitivity_array[rowNumber-1][columnNumber-1][0] in rank0_str:
                    color = 'papaya whip'
                elif transitivity_array[rowNumber-1][columnNumber-1] in rank1_str:
                    color = 'lightskyblue1'
                rect = view.create_rectangle(col * gridWidth,row * gridHeight,(col + 1) * gridWidth,(row + 1) * gridHeight,fill = color,outline='white')
                rect_text= transitivity_array[rowNumber-1][columnNumber-1]
                if len(list(rect_text)) == 13:
                    rect_text = "All \nrelations"
                elif(len(list(rect_text)) > 4):
                    rect_text = (str(list(rect_text)[0:4]) + '\n' + str(list(rect_text)[4:])).replace(',','').replace('\'','').replace('[','').replace(']','')
            view.itemconfig(rect, tags=(str(rowNumber), str(columnNumber)))
            t = view.create_text(col * gridWidth + 50,row * gridHeight+20, text=rect_text, font=("Courier", 13))
            view.pack()
            
def get_row_and_column_from_grid(string_info):
    for i,t1 in enumerate(transitivity_array):
        for j,t2 in enumerate(t1):
            if t2 == string_info:
                if i!=0 and j!=0:
                    return (i+1,j+1)
    return (0,0)
            
def draw_rank2(view, selected, transitive):
    index = 0
    for rank2 in rank2_str:
        if len(rank2) > 6:
            disp_string_r2 = (str(rank2[0:6])+"\n"+str(rank2[6:])).replace('\'','').replace('[','').replace(']','').replace(',','')
        else:
            disp_string_r2 = str(rank2).replace('\'','').replace('[','').replace(']','').replace(',','')
        if index == selected:
            color = "medium spring green"
            border_color = "black"
        elif index == transitive:
            color = "tomato"
            border_color = "white"
        else:
            color = "MistyRose2"
            border_color = "white"
        rect = view.create_rectangle(4+135*index,605,3+135*index+135,605+50,fill =color,outline=border_color, width=2)
        rect_text= disp_string_r2
        t = view.create_text(3+135*index + 70,605+20, text=rect_text, font=("Courier", 13))
        view.itemconfig(rect, tags=index)

        index = index+1
            
'''
Clears the given frame
'''
def clear_frame(frame):
    for widget in frame.winfo_children():
        widget.destroy()            

'''
Showing the string representation in the frame 
'''        
def show_s_word(root, word):
    if sword_fr is not None:
        clear_frame(sword_fr)
    for i, s in enumerate(word):
        display_str = str(s).replace('{', '').replace('}', '').replace('\'', '')
        f = Frame(root, height=30, width=70)
        f.pack_propagate(0) 
        f.place(x=5+70*i, y=10)
        label = Label(f, text=display_str, borderwidth=2, relief="ridge", bg='lavender')  #label
        label.pack(fill=BOTH, expand=1)

'''
Gets rank of a particular string
'''
def get_rank(temp):
    if len(temp) == 1 and temp[0] in rank0_str:
        rank = 0
    elif temp in rank1_str:
        rank = 1
    elif temp in rank2_str:
        rank = 2
    else:
        rank = 3
    return rank

'''
Gets the display value particular string
'''
def get_display_string(temp):
    if len(temp) > 6:
        disp_string = (str(temp[0:6])+"\n"+str(temp[6:])).replace('\'','').replace('[','').replace(']','')
    else:
        disp_string = str(temp).replace('\'','').replace('[','').replace(']','')
    return disp_string
    
'''
Click hanlder on the transitivity table
'''    
def click_on_box(event):
    global previous,s_words_info_label, fr, view, sword_fr,result,choice,len_n,length_fr,prob_fr,prob_choice_fr,first_clicked_relation,is_second_clicked,trans_f,transitivity_rel_set,first_clicked_row,first_clicked_column,rank_info_label_text
    choice.set(0)
    len_n.set(0)
    s_words_info_label['text'] = " "
    prob_info_label['text'] = " "
    if transitivity_rel_set:
        transitivity_info_label_text.set("")
        transitivity_rel_set = False
    view.pack()
    if view.find_withtag(CURRENT):
        if len(view.gettags(CURRENT)) == 3:
            (row, column, tag) = view.gettags(CURRENT)
            if is_second_clicked:
                if(int(row) != 1 and int(column) != 1):
                    second_clicked_relation = transitivity_array[int(row)-1][int(column)-1]
                    temp1 = []
                    for r1 in first_clicked_relation:
                        for r2 in second_clicked_relation:
                            relation,s = transitivity_relation(r1,r2)
                            temp1.extend(relation)
                    temp1.sort()       
                    temp = list(set(temp1)) 
                    temp.sort()

                    rank = get_rank(temp)
                    disp_string = get_display_string(temp)
                    print("Transitivity relation between " + str(first_clicked_relation) + " and " + str(second_clicked_relation)  + " is : ", str(temp), "of rank ", rank)
                    print("--------------------")
                    if rank != 3:
                        transitivity_info_label_text.set("Transitivity relation: "+ disp_string)
                        rank_info_label_text.set("Rank: "+ str(rank))
                        transitivity_rel_set = True

                    shorter_word = get_shorter_s_word(temp)
                    sword_fr = Frame(fr, height=60, width=300)
                    sword_fr.pack_propagate(0) 
                    sword_fr.place(x=0, y=60)

                    if shorter_word == None or len(shorter_word) == 0:
                        if(len(temp) == 1):
                            s_words_info_label_text.set("Shortest representation:")
                            shorter_word = allen('i','j',temp[0])
                            show_s_word(sword_fr, shorter_word)
                        else:
                            shorter_word = "Not available"
                            s_words_info_label_text.set("Shortest representation: "+ str(shorter_word))
                            clear_frame(sword_fr)
                            i = 0
                            while(i<len(other_representations)):
                                if(set(temp) == set(other_representations[i])):
                                    to_show = other_rep_s_words[i]
                                    show_s_word(sword_fr, to_show)
                                i = i+1
                    else:
                        s_words_info_label_text.set("Shortest representation: ")
                        show_s_word(sword_fr, shorter_word)

                    if rank == 1:
                        rel_row_column = get_row_and_column_from_grid(temp)
                        draw_grid(view, (first_clicked_row, first_clicked_column), (int(row), int(column)), rel_row_column, (None, None))
                    elif rank == 2:
                        for index, rank2_s in enumerate(rank2_str):
                            if temp == rank2_s:
                                to_color = index
                                break
                        draw_grid(view, (first_clicked_row, first_clicked_column), (int(row), int(column)), (None, None), (None, None))
                        draw_rank2(view, None, to_color)
                is_second_clicked = False
                
            else:
                if(int(row) != 1 and int(column) != 1):
                    first_clicked_row = int(row)
                    first_clicked_column = int(column)
                    is_second_clicked = True
                    first_clicked_relation = transitivity_array[int(row)-1][int(column)-1]
                    shorter_word = get_shorter_s_word(transitivity_array[int(row)-1][int(column)-1])
                    sword_fr = Frame(fr, height=60, width=300)
                    sword_fr.pack_propagate(0) 
                    sword_fr.place(x=0, y=60)
                    if len(first_clicked_relation) == 1:
                        rank = 0
                    else:
                        rank = 1
                    transitivity_info_label_text.set("Transitivity relation: - ")
                    rank_info_label_text.set("Rank: "+ str(rank))
                    if shorter_word == None or len(shorter_word) == 0:
                        if(len(first_clicked_relation) == 1):
                            s_words_info_label_text.set("Shortest representation: ")
                            shorter_word = allen('i','j',first_clicked_relation[0])
                            show_s_word(sword_fr, shorter_word)
                        else:
                            shorter_word = "Not available"
                            s_words_info_label_text.set("Shortest representation: "+ str(shorter_word))
                            clear_frame(sword_fr)
                            i = 0
                            while(i<len(other_representations)):
                                if(set(first_clicked_relation) == set(other_representations[i])):
                                    to_show = other_rep_s_words[i]
                                    show_s_word(sword_fr, to_show)
                                i = i+1
                    else:
                        s_words_info_label_text.set("Shortest representation: ")
                        show_s_word(sword_fr, shorter_word)

                    if rank == 1:
                        for r1 in rank1_dict:
                            if r1.get('string') == first_clicked_relation:
                                transitive_relations = r1.get('relation')
                                break
                        rel1_row_column = get_row_and_column_from_grid([transitive_relations[0]])
                        rel2_row_column = get_row_and_column_from_grid([transitive_relations[1]])
                        print("The selected box ("+ str(first_clicked_relation) +") can be obtained from : "+ str(transitive_relations[0]) + " and "+ str(transitive_relations[1]))
                    else:
                        rel1_row_column = (None, None)
                        rel2_row_column = (None, None)
                        print("The selected box ("+ str(first_clicked_relation) +") is of rank 0")
                    print("--------------------")
                    draw_grid(view, (int(row), int(column)), (None, None), rel1_row_column, rel2_row_column)
                    draw_rank2(view, None, None)
                    previous = view.gettags(CURRENT)
                             
        elif len(view.gettags(CURRENT)) == 2:
            (rank2_i,tag) = view.gettags(CURRENT)
            rank2_index = int(rank2_i)
            view.itemconfig(CURRENT, fill="medium spring green",outline='black', width=2)
            if is_second_clicked:
                second_clicked_relation = rank2_str[rank2_index]
                temp1 = []
                for r1 in first_clicked_relation:
                    for r2 in second_clicked_relation:
                        relation,s = transitivity_relation(r1,r2)
                        temp1.extend(relation)
                temp1.sort()       
                temp = list(set(temp1)) 
                temp.sort()
                rank = get_rank(temp)
                disp_string = get_display_string(temp)
                print("Transitivity relation between " + str(first_clicked_relation) + " and " + str(second_clicked_relation)  + " is : ", str(temp), "of rank ", rank)
                print("--------------------")
                if rank != 3:
                    transitivity_info_label_text.set("Transitivity relation: "+ disp_string)
                    rank_info_label_text.set("Rank: "+ str(rank))
                    transitivity_rel_set = True

                shorter_word = get_shorter_s_word(temp)
                sword_fr = Frame(fr, height=60, width=300)
                sword_fr.pack_propagate(0) 
                sword_fr.place(x=0, y=60)

                if shorter_word == None or len(shorter_word) == 0:
                    if(len(temp) == 1):
                        s_words_info_label_text.set("Shortest representation:")
                        shorter_word = allen('i','j',temp[0])
                        show_s_word(sword_fr, shorter_word)
                    else:
                        shorter_word = "Not available"
                        s_words_info_label_text.set("Shortest representation: "+ str(shorter_word))
                        clear_frame(sword_fr)
                        i = 0
                        while(i<len(other_representations)):
                            if(set(temp) == set(other_representations[i])):
                                to_show = other_rep_s_words[i]
                                show_s_word(sword_fr, to_show)
                            i = i+1
                else:
                    s_words_info_label_text.set("Shortest representation: ")
                    show_s_word(sword_fr, shorter_word)

                if rank == 1:
                    if len(temp) == 13:
                        rel_row_column = (2,3)
                    else:
                        rel_row_column = get_row_and_column_from_grid(temp)
                    draw_grid(view, (first_clicked_row, first_clicked_column), (None, None), rel_row_column, (None, None))
                else:
                    for index, rank2_s in enumerate(rank2_str):
                        if temp == rank2_s:
                            to_color = index
                            break
                    draw_grid(view, (first_clicked_row, first_clicked_column), (None, None), (None, None), (None, None))
                    draw_rank2(view, None, to_color)
                is_second_clicked = False
            else:
                is_second_clicked = True
                first_clicked_row = None
                first_clicked_column = None
                first_clicked_relation = rank2_str[rank2_index]
                shorter_word = get_shorter_s_word(rank2_str[rank2_index])
                sword_fr = Frame(fr, height=60, width=300)
                sword_fr.pack_propagate(0) 
                sword_fr.place(x=0, y=60)
                rank = 2
                transitivity_info_label_text.set("Transitivity relation: - ")
                rank_info_label_text.set("Rank: "+ str(rank))
                if shorter_word == None or len(shorter_word) == 0:
                    if(len(first_clicked_relation) == 1):
                        s_words_info_label_text.set("Shortest representation: ")
                        shorter_word = allen('i','j',first_clicked_relation[0])
                        show_s_word(sword_fr, shorter_word)
                    else:
                        shorter_word = "Not available"
                        s_words_info_label_text.set("Shortest representation: "+ str(shorter_word))
                        clear_frame(sword_fr)
                        i = 0
                        while(i<len(other_representations)):
                            if(set(first_clicked_relation) == set(other_representations[i])):
                                to_show = other_rep_s_words[i]
                                show_s_word(sword_fr, to_show)
                            i = i+1
                else:
                    s_words_info_label_text.set("Shortest representation: ")
                    show_s_word(sword_fr, shorter_word)
                if rank == 2:
                    for r1 in rank2_dict:
                        if r1.get('string') == first_clicked_relation:
                            transitive_relations = r1.get('relation')
                            break
                rel1_row_column = get_row_and_column_from_grid(transitive_relations[0])
                rel2_row_column = get_row_and_column_from_grid(transitive_relations[1])
                print("The selected box ("+ str(first_clicked_relation) +") can be obtained from : "+ str(transitive_relations[0]) + " and "+ str(transitive_relations[1]))
                print("--------------------")
                draw_grid(view, (None, None), (None, None), rel1_row_column, rel2_row_column)
                draw_rank2(view, rank2_index, None)
                previous = view.gettags(CURRENT)
        

'''
Clear Frame
'''
def clear_canvas(frame):
    for widget in frame.winfo_children():
        widget.destroy()

'''
Initializing tkinter and canvas
'''
root = Tk()
root.geometry("1470x810+5+5")
root.title("Allen Relations")
view = Canvas(root, width=1400, height=700)
view.pack(side=TOP, fill=BOTH,padx=3,pady=3)
all_relations = ['b','bi','d','di','eq','f','fi','m','mi','o','oi','s','si']
possible_short_representations = [[{'li'},{'lj'}],     [{'li'},{'rj'}],  
                            [{'lj'},{'li'}],     [{'rj'},{'li'}],
                            [{'li', 'lj'}],      [{'li', 'rj'}],  
                            [{'ri'},{'lj'}],     [{'ri'},{'rj'}],  
                            [{'lj'},{'ri'}],     [{'rj'},{'ri'}],
                            [{'ri','lj'}],     [{'ri','rj'}],  
                            [{'lj'},{'ri'},{'rj'}],
                            [{'lj'},{'li'},{'rj'}],
                            [{'li'},{'rj'},{'ri'}],
                            [{'li'},{'lj'},{'ri'}]]
other_representations = [['b', 'm', 'o'], ['bi','mi','oi'], ['d','di','eq','f','fi','o','oi','s','si']]
other_rep_s_words = [[{'li'},{'<ri>,<lj>'},{'rj'}], [{'lj'},{'<rj>,<li>'},{'ri'}], [{'<li>, <lj>'},{'<ri>, <rj>'}]]
previous = None
is_second_clicked = False
first_clicked_relation = None
transitivity_rel_set = False
rank0_str = rank0()
rank1_str,rank1_dict = rank1()
rank2_str,rank2_dict = rank2()
rank3_str = rank3()
result=[]
superpose_result=[]
choice=IntVar()
choice.set(0)
len_n=IntVar()
len_n.set(0)
style = tkFont.Font(family='Courier', size=14)
transitivity_array = compute_transitivty_table()
shorter_representations = possible_s_words()
draw_grid(view, (None, None), (None, None), (None, None), (None, None))
draw_rank2(view, None, None)


'''
Adding the frame to show string representation and probability information
'''
fr = Frame(root, height=200, width=2000)
fr.pack_propagate(0) # don't shrink
fr.place(x=5, y=660)

'''
Adding the string representation information
'''
Label(view, text="* Click on the grid elements to see string representation and transitivity information", font=("Courier", 12)).place(x=800, y=607)
Label(view, text="Legends", font=("Courier", 12)).place(x=290, y=605)
Label(view, text="Rank0", font=("Courier", 12), bg="papaya whip", borderwidth=2, relief="flat").place(x=350, y=607)
Label(view, text="Rank1", font=("Courier", 12), bg="LightSteelBlue1", borderwidth=2, relief="flat").place(x=400, y=607)
Label(view, text="Rank2", font=("Courier", 12), bg="MistyRose2", borderwidth=2, relief="flat").place(x=450, y=607)
Label(view, text="Selected", font=("Courier", 12), bg="medium spring green", borderwidth=2, relief="solid").place(x=500, y=607)
Label(view, text="Transitivity Info", font=("Courier", 12), bg="tomato", borderwidth=2, relief="flat").place(x=570, y=607)
s_words_info_label_text = StringVar()
s_words_info_label = Label(fr, textvariable=s_words_info_label_text, font=("Courier", 13))
s_words_info_label.place(x=0, y=40)

transitivity_info_label_text = StringVar()
transitivity_info_label = Label(fr, textvariable=transitivity_info_label_text, font=("Courier", 13))
transitivity_info_label.place(x=0, y=0)

rank_info_label_text = StringVar()
rank_info_label = Label(fr, textvariable=rank_info_label_text, font=("Courier", 13))
rank_info_label.place(x=0, y=20)

view.bind("<Button-1>", click_on_box)

'''
Adding the probablity information
'''
prob_info_label_text = StringVar()
prob_info_label = Label(fr, textvariable=prob_info_label_text, font=("Courier", 13))
prob_info_label.place(x=450, y=0)

root.mainloop()
```
